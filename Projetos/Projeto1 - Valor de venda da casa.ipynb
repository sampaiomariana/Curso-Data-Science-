Pré-Processamento dos dados

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler, OneHotEncoder
```

```python
x
```

```python
min_max = MinMaxScaler()
scaled_x = min_max.fit_transform(x.drop(['OverallQual'], axis = 1))
scaled_x[0]
```

```python
#É um transformador da variável que for colocada nele.
#Tem fi e fit_transformer
one = OneHotEncoder(categories = 'auto')
qual_trans = one.fit_transform(x[['OverallQual']])
qual_trans.todense()
```

```python
#Categoria 7 nota 1
one.categories_
```

```python
#Cada nota tem um peso especifico 
#Codificado
qual_trans[0].todense()
```

```python
qual_trans = pd.DataFrame(qual_trans.todense(),columns=one.categories_)
qual_trans.columns = ['qual_'+str(x) for x in list(zip(*qual_trans.columns.values))[0]]
```

```python
df_scaled = pd.DataFrame(scaled_x,columns=x.drop(['OverallQual'],axis=1).columns)
df_scaled
```

```python
for column in qual_trans.columns:
    df_scaled[column] = qual_trans[column]
```

```python
df_scaled
```

```python
x_train,x_test,y_train,y_test = train_test_split(df_scaled,y, test_size = 0.25, random_state=43)
```

```python
lr_scaled = LinearRegression()
lr_scaled.fit(x_train,y_train)
```

```python
#Diminuir o erro em 10% é bom
#Utilizar mais features pode levar a redução 
y_pred = lr_scaled.predict(x_test)
np.sqrt(mean_squared_error(y_test,y_pred))
```
